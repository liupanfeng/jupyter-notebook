```python
from lxml.html import parse
from urllib.request import urlopen
```

```python
parsed=parse(urlopen('https://ask.julyedu.com/'))
doc=parsed.getroot()
```

```python
doc
```

```python
links=doc.findall('.//a')
```

```python
links[15:20]
type(links)
```

```python
link=links[14]
link
link.get('herf')
print(link.text_content())
```

```python
urls=[lnk.get('href') for lnk in doc.findall('.//a')]#获取所有的链接
urls[-10:]
```

```python
spans=doc.findall('.//span')
len(spans)
```

```python
def _unpack(spans):
    return [val.text_content() for val in spans]
```

```python
contents=_unpack(spans)
for content in contents:
    print (content)
```

```python
questions=doc.findall('.//h4')
len(questions)
```

```python
contents=_unpack(questions)
for content in contents:
    print(content)
```

解析html

```python
from lxml import objectify

path = 'Performance_MNR.xml'
parsed = objectify.parse(open(path))
root = parsed.getroot()
```

```python
data=[]
skip_fields=['PARENT_SEQ', 'INDICATOR_SEQ',
               'DESIRED_CHANGE', 'DECIMAL_PLACES']
for elt in root.INDICATOR:
    el_data={}
    for child in elt.getchildren():
        print("tag---",child.tag)
        if child.tag in skip_fields:
            continue
        print("pyval--",child.pyval)
        el_data[child.tag] = child.pyval
    data.append(el_data)
```

```python
pref=DataFrame(data)
pref
```
