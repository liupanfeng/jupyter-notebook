```python
import requests
result=requests.get("http://news.qq.com/")
print(result.status_code) #可以通过状态码来确定是否成功
```

```python
def get_status(url):
    r = requests.get(url, allow_redirects = False)#allow_redirects  表示允许302跳转
    return r.status_code

print (get_status('http://www.zhidaow.com'))
#200
print (get_status('http://www.zhidaow.com/hi404/'))
#404
print (get_status('http://www.baidu.com/link?url=QeTRFOS7TuUQRppa0wlTJJr6FfIYI1DJprJukx4Qy0XnsDO_s9baoO8u1wvjxgqN'))
#302
```

```python
result.encoding
```

可以看到网页的编码格式为GB2312

从返回的结果result中，可以取出网页的内容

```python
content=result.content
print(content[:2000])#获取前1000个字符
```

以上为网页的HTML格式数据，接着我们就可以从上面取出所需要的内容了，比如现在我们需要爬取的是这一页腾讯新闻的新闻标题。
我们要用到另外一个非常好用的库，叫做BeautifulSoup，我们用它完成网页返回结果的解析和内容的提取。

具体的BeautifulSoup使用方法可以参考它的中文文档

```python
import warnings
warnings.filterwarnings("ignore")

from bs4 import BeautifulSoup
soup=BeautifulSoup(content,fromEncoding="GB2312")
sample=soup.find_all("a","linkto") #前面是标签名 后面是class名字  
sample[0]
```

我们需要标题，可以直接从sample中取出来，通过sample.string完成这个操作

```python
print(sample[0].string)
```

如果我们想把所有的标题取出来，只需要将刚才的做个简单的修改就行了

```python
[sam.string for sam in sample]#这样就能取出所有的标题内容了
```

再URL中传递参数

有时候我们需要在url中传递参数，比如在采集百度搜索结果时，我们wd参数（搜索词）和rn参数（搜索结果数量）
可以手工组成一个url

```python
payload = {'wd': '七月在线数据分析班', 'rn': '10'}
r=requests.get("http://www.baidu.com/s",params=payload)
print(r.url)
```

```python
r.content
```

设置超时时间  可以通过timeout属性来设置超时时间，一旦超过这个时间还没有获取数据就提示错误

```python
requests.get('http://github.com',timeout=1)
```

添加代理  爬虫爬去数据时为了避免被封ip，经常会使用代理。requests也有相应的属性。


```python
proxies = {
  "http": "http://10.10.1.10:3128",
  "https": "http://10.10.1.10:1080",
}

requests.get("http://www.zhidaow.com", proxies=proxies)
```

```python
proxies = {
    "http": "http://user:pass@10.10.1.10:3128/",#如果代理需要账户和密码
}
```

2.8 模拟登陆
有很多网站需要我们登录之后才能完成数据的抓取，requests同样支持提交表单(包含用户名和密码)进行登录。代码的示例如下：

```python
s = requests.session()
data = {'user':'用户名','passdw':'密码'}
#post 换成登录的地址，
res=s.post('http://www.xxx.net/index.php?action=login',data);
#换成抓取的地址
s.get('http://www.xxx.net/archives/155/');
```
